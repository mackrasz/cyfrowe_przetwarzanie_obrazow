```
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def load_image(filename):
  im = np.array(Image.open(filename)).mean(axis=2)/255
  return im

def draw_image(im):
  plt.figure(figsize=(12,9))
  plt.imshow(im, cmap='gray')
  plt.clim(0, 1)
  plt.colorbar()
```
